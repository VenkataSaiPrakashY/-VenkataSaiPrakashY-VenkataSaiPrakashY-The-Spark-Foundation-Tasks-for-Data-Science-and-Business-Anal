```python
import pandas as pd
data = pd.read_csv('AmesHousing.csv')
data.describe()
```

```python
data.describe()
```

```python
data.columns
```

```python
data.head(10)
```

```python
data.dropna(axis=0)
data.describe()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
model = DecisionTreeRegressor()
x = data[['Lot Area', 'Year Built', '1st Flr SF', '2nd Flr SF', 'Full Bath', 'Bedroom AbvGr', 'TotRms AbvGrd' ]]
y = data['SalePrice']
x_train, xval, ytrain, yval = train_test_split(x,y,random_state=0)
```

```python
model.fit(x_train, ytrain)
```

```python
result = model.predict(xval.head())
print(result)
print(yval.head())
```
